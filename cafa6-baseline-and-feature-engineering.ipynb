```python
def load_dataset_tensorflow_train(embeds, go_terms):
    return {"embeds": tf.cast(embeds, tf.float32), "go_terms": tf.cast(go_terms, tf.float32)}

def dict_to_tuple(inputs):
    return inputs["embeds"], inputs["go_terms"]
```

```python
def generate_tf_datasets(p_df, p_BATCH_SIZE_PER_REPLICA, p_label_cols):

    BATCH_SIZE = p_BATCH_SIZE_PER_REPLICA * tpu_strategy.num_replicas_in_sync
    
    embeds = pl.Series(p_df.select(pl.col('protein_embedding'))).to_list()
    go_terms = pl.Series(p_df.select(p_label_cols)).map_elements(lambda d: list(d.values())).to_list()

    """num_samples = p_df.shape[0]

    # Use NumPy tiling or broadcasting to repeat the 1D weights for every sample
    tiles = np.tile(p_wts, (num_samples, 1))
    label_wts = pl.Series(tiles).to_list()"""
    
    tf_dataset = tf.data.Dataset.from_tensor_slices((embeds, go_terms))
    
    tf_ds = tf_dataset.map(load_dataset_tensorflow_train, num_parallel_calls=tf.data.AUTOTUNE)
    tf_ds = tf_ds.map(dict_to_tuple, num_parallel_calls=tf.data.AUTOTUNE)
    tf_ds = tf_ds.batch(batch_size=BATCH_SIZE, drop_remainder=True)
    tf_ds = tf_ds.prefetch(tf.data.AUTOTUNE)
    
    return tf_ds
```

```python
pos_dict = {}
for i in range(len(label_cols_cc)):
    pos_dict[i] = label_cols_cc[i]

print(pos_dict[0])
```

```python
def load_dataset_tensorflow_pred(proteins, embeds):
    return {"proteins": proteins,"embeds": tf.cast(embeds, tf.float32)}

def dict_to_tuple_pred(inputs):
    return inputs["proteins"], inputs["embeds"]

def generate_tf_datasets_pred(p_df, p_BATCH_SIZE_PER_REPLICA):

    BATCH_SIZE = p_BATCH_SIZE_PER_REPLICA * tpu_strategy.num_replicas_in_sync
    
    embeds = pl.Series(p_df.select(pl.col('embedding_arrays'))).to_list()
    proteins = pl.Series(p_df.select(pl.col('protein_accession_id'))).to_list()
    
    tf_dataset = tf.data.Dataset.from_tensor_slices((proteins, embeds))
    
    tf_ds = tf_dataset.map(load_dataset_tensorflow_pred, num_parallel_calls=tf.data.AUTOTUNE)
    tf_ds = tf_ds.map(dict_to_tuple_pred, num_parallel_calls=tf.data.AUTOTUNE)
    tf_ds = tf_ds.batch(batch_size=BATCH_SIZE, drop_remainder=False)
    tf_ds = tf_ds.prefetch(tf.data.AUTOTUNE)
    
    return tf_ds
```

```python
pred_ds = generate_tf_datasets_pred(p_df=df_test_features, p_BATCH_SIZE_PER_REPLICA = 16)
```
