```python
def load_dataset_tensorflow_train(embeds, go_terms):
    return {"embeds": tf.cast(embeds, tf.float32), "go_terms": tf.cast(go_terms, tf.float32)}

def dict_to_tuple(inputs):
    return inputs["embeds"], inputs["go_terms"]
```

```python
def generate_tf_datasets(p_df, p_BATCH_SIZE_PER_REPLICA):

    BATCH_SIZE = p_BATCH_SIZE_PER_REPLICA * tpu_strategy.num_replicas_in_sync
    
    embeds = pl.Series(p_df.select(pl.col('uniprot_embeds'))).to_list()
    go_terms = pl.Series(p_df.select(label_cols)).map_elements(lambda d: list(d.values())).to_list()
    
    tf_dataset = tf.data.Dataset.from_tensor_slices((embeds, go_terms))
    
    tf_ds = tf_dataset.map(load_dataset_tensorflow_train, num_parallel_calls=tf.data.AUTOTUNE)
    tf_ds = tf_ds.map(dict_to_tuple, num_parallel_calls=tf.data.AUTOTUNE)
    tf_ds = tf_ds.batch(batch_size=BATCH_SIZE, drop_remainder=True)
    tf_ds = tf_ds.prefetch(tf.data.AUTOTUNE)
    
    return tf_ds
```

```python
train_ds = generate_tf_datasets(p_df=x_train, p_BATCH_SIZE_PER_REPLICA = 16)
test_ds = generate_tf_datasets(p_df=x_test, p_BATCH_SIZE_PER_REPLICA = 16)
valid_ds = generate_tf_datasets(p_df=x_valid, p_BATCH_SIZE_PER_REPLICA = 16)
```

```python
from tensorflow.keras import layers

with tpu_strategy.scope():
    
    protein_input = layers.Input(shape=(1024,), name='protein_input')
    h1 = layers.Dense(1024, activation='selu', name='hidden_1', kernel_initializer='lecun_normal')(protein_input)
    d1 = layers.AlphaDropout(0.1, name='dropout_1')(h1)
    h2 = layers.Dense(1000, activation='selu', kernel_initializer='lecun_normal')(d1)
    d2 = layers.AlphaDropout(0.2, name='dropout_2')(h2)
    h3 = layers.Dense(800, activation='selu', kernel_initializer='lecun_normal')(d2)
    d3 = layers.AlphaDropout(0.3, name='dropout_3')(h3)
    h4 = layers.Dense(650, activation='selu', kernel_initializer='lecun_normal')(h3)
    h5 = layers.Dense(450, activation='selu', kernel_initializer='lecun_normal')(h4)
    h6 = layers.Dense(250, activation='selu', kernel_initializer='lecun_normal')(h5)
    protein_output = layers.Dense(1500, activation="sigmoid")(h6)
    
    model = tf.keras.Model(protein_input, protein_output)

    checkpoint_cb = tf.keras.callbacks.ModelCheckpoint("keras_baseline.keras")
    #early_stopping_cb = tf.keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)
    tensor_board_cb = tf.keras.callbacks.TensorBoard(log_dir='/kaggle/working/logs', histogram_freq=1)

    callbacks_list = [checkpoint_cb, tensor_board_cb]

    model_optimizer = tf.keras.optimizers.Adamax(
        learning_rate=0.00001
        #,clipnorm=1.0
    )
    
    """model.compile(loss='binary_crossentropy', optimizer=model_optimizer
                   , metrics=['binary_accuracy'])"""

    """model.compile(loss=tf.keras.losses.CategoricalCrossentropy(from_logits=False), optimizer=model_optimizer
                   , metrics=[tf.keras.metrics.AUC(multi_label=True, num_labels=1500, from_logits=False
                                                   , label_weights=label_wts)])"""
    model.compile(loss=tf.keras.losses.BinaryCrossentropy(from_logits=False), optimizer=model_optimizer
                   , metrics=[tf.keras.metrics.AUC(multi_label=True, num_labels=1500, from_logits=False
                                                   , label_weights=label_wts)])
```

```python
pos_dict = {}
for i in range(len(label_cols)):
    pos_dict[i] = label_cols[i]

print(pos_dict[0])
```

```python
def load_dataset_tensorflow_pred(proteins, embeds):
    return {"proteins": proteins,"embeds": tf.cast(embeds, tf.float32)}

def dict_to_tuple_pred(inputs):
    return inputs["proteins"], inputs["embeds"]

def generate_tf_datasets_pred(p_df, p_BATCH_SIZE_PER_REPLICA):

    BATCH_SIZE = p_BATCH_SIZE_PER_REPLICA * tpu_strategy.num_replicas_in_sync
    
    embeds = pl.Series(p_df.select(pl.col('uniprot_embeds'))).to_list()
    proteins = pl.Series(p_df.select(pl.col('protein_accession_id'))).to_list()
    
    tf_dataset = tf.data.Dataset.from_tensor_slices((proteins, embeds))
    
    tf_ds = tf_dataset.map(load_dataset_tensorflow_pred, num_parallel_calls=tf.data.AUTOTUNE)
    tf_ds = tf_ds.map(dict_to_tuple_pred, num_parallel_calls=tf.data.AUTOTUNE)
    tf_ds = tf_ds.batch(batch_size=BATCH_SIZE, drop_remainder=False)
    tf_ds = tf_ds.prefetch(tf.data.AUTOTUNE)
    
    return tf_ds
```

```python
pred_ds = generate_tf_datasets_pred(p_df=df_test_features, p_BATCH_SIZE_PER_REPLICA = 16)
```

```python
all_predictions = []
for proteins, embeds in pred_ds:
    #print(f"Images: {dicoms.numpy().shape}, DCM Paths: {dcm_paths.shape}")
    batch_predictions = model.predict(embeds)
    for prots, pred in zip(proteins, batch_predictions):
        #print("path is -",path)
        pred_copy = pred.copy().reshape(1, -1)
        #pred_copy[0, 7] = 1 - pred_copy[0, 7]
        data_dict = {'protein': prots.numpy().decode('utf-8')}
        data_dict.update({pos_dict[i]: float(pred) for i, pred in enumerate(pred_copy[0])})
        all_predictions.append(data_dict)
```
