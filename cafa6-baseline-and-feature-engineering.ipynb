```python
def load_dataset_tensorflow_train(embeds, go_terms):
    return {"embeds": tf.cast(embeds, tf.float32), "go_terms": tf.cast(go_terms, tf.float32)}

def dict_to_tuple(inputs):
    return inputs["embeds"], inputs["go_terms"]
```

```python
def generate_tf_datasets(p_df, p_BATCH_SIZE_PER_REPLICA):

    BATCH_SIZE = p_BATCH_SIZE_PER_REPLICA * tpu_strategy.num_replicas_in_sync
    
    embeds = pl.Series(p_df.select(pl.col('uniprot_embeds'))).to_list()
    go_terms = pl.Series(p_df.select(label_cols)).map_elements(lambda d: list(d.values())).to_list()
    
    tf_dataset = tf.data.Dataset.from_tensor_slices((embeds, go_terms))
    
    tf_ds = tf_dataset.map(load_dataset_tensorflow_train, num_parallel_calls=tf.data.AUTOTUNE)
    tf_ds = tf_ds.map(dict_to_tuple, num_parallel_calls=tf.data.AUTOTUNE)
    tf_ds = tf_ds.batch(batch_size=BATCH_SIZE, drop_remainder=True)
    tf_ds = tf_ds.prefetch(tf.data.AUTOTUNE)
    
    return tf_ds
```

```python
train_ds = generate_tf_datasets(p_df=x_train, p_BATCH_SIZE_PER_REPLICA = 8)
test_ds = generate_tf_datasets(p_df=x_test, p_BATCH_SIZE_PER_REPLICA = 8)
valid_ds = generate_tf_datasets(p_df=x_valid, p_BATCH_SIZE_PER_REPLICA = 8)
```
